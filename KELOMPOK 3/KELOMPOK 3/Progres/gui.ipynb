```python
import tkinter as tk
from sklearn.neighbors import KNeighborsClassifier
from joblib import load
from PIL import Image, ImageTk

def calculate_aqi():
    loaded_model = load('C:/Users/RAKHA ALCANDER/OneDrive/Desktop/AI/Progres gui/Progres/model_GA.joblib')

    # Mendapatkan nilai dari entry fields
    entry_aqi1 = float(entry_aqi.get())
    ozone_value = float(entry_ozone.get())
    no2_value = float(entry_no2.get())
    no2_cat = float(entry_no2_cat.get())
    pm25_value = float(entry_pm25.get())
    pm25_cat = float(entry_pm25_cat.get())

    input_data = [[entry_aqi1, ozone_value, no2_value, no2_cat, pm25_value, pm25_cat]]

    # Lakukan prediksi dengan model KNN yang dimuat
    prediction = loaded_model.predict(input_data)

    # Menampilkan hasil prediksi pada entry field hasil
    hasil_cat.delete(0, tk.END)
    hasil_cat.insert(0, str(prediction))
    
root = tk.Tk()
root.title("AQI Calculator")
root.geometry("720x400")

image = Image.open("C:/Users/RAKHA ALCANDER/OneDrive/Desktop/AI/Progres gui/Progres/mencoba1.png")
background_image = ImageTk.PhotoImage(image)

image_label = tk.Label(root, image=background_image)
image_label.place(x=0,y=0)


# Label dan Entry untuk nilai AQI
label_aqi = tk.Label(root, text="AQI Value")
label_aqi.place(x=240, y=120)

entry_aqi = tk.Entry(root)
entry_aqi.place(x=400, y=120)

# Label dan Entry untuk nilai Ozone AQI Category
label_ozone = tk.Label(root, text="Ozone AQI Category")
label_ozone.place(x=240, y=150)

entry_ozone = tk.Entry(root)
entry_ozone.place(x=400, y=150)

# Label dan Entry untuk nilai NO2 AQI Value
label_no2 = tk.Label(root, text="NO2 AQI Value")
label_no2.place(x=240, y=180)

entry_no2 = tk.Entry(root)
entry_no2.place(x=400, y=180)

# Label dan Entry untuk nilai NO2 AQI Category
label_no2_cat = tk.Label(root, text="NO2 AQI Category")
label_no2_cat.place(x=240, y=210)

entry_no2_cat = tk.Entry(root)
entry_no2_cat.place(x=400, y=210)

# Label dan Entry untuk nilai PM2.5 AQI Value
label_pm25 = tk.Label(root, text="PM2.5 AQI Value")
label_pm25.place(x=240, y=240)

entry_pm25 = tk.Entry(root)
entry_pm25.place(x=400, y=240)

# Label dan Entry untuk nilai PM2.5 AQI Category
label_pm25_cat = tk.Label(root, text="PM2.5 AQI Category")
label_pm25_cat.place(x=240, y=270)

entry_pm25_cat = tk.Entry(root)
entry_pm25_cat.place(x=400, y=270)

# Entry untuk menampilkan hasil prediksi AQI
hasil_cat = tk.Entry(root)
hasil_cat.place(x=400, y=300)

# Tombol untuk menghitung AQI
calculate_button = tk.Button(root, text="Calculate AQI", command=calculate_aqi)
calculate_button.place(x=240, y=300)

root.mainloop()
```
