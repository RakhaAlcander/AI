```python
df_reg = df.iloc[:,[-5,-9,-10,1]]
df_reg.head()
```

```python
new_column_name = {df_reg.columns[0]: "A",df_reg.columns[1]:'B',df_reg.columns[2]:'C',df_reg.columns[3]:'label'}
df_reg = df_reg.rename(columns=new_column_name)
df_reg.head()
```

```python
import math
import time
import pandas as pd
from collections import Counter

def akurasi_score(y_test, y_pred):
    jumlah_benar = 0
    for prediksi, label in zip(y_pred, y_test):
        if prediksi == label:
            jumlah_benar += 1

    akurasi = jumlah_benar / len(y_test)
    return akurasi

def knn_predict(data,tetangga):
  y = data
  z = tetangga
  hasil_prediksi = []

  for data in y:
    if z == 1:
      x = jarak_euclidean(data, 1)  # Menghitung jarak terkecil
      prediksi = df_reg['label'][x]  # Menggunakan jarak terkecil
      hasil_prediksi.append(prediksi)

    else:
      x_multiple = jarak_euclidean(data, z)
      label_multiple = [df_reg['label'][i] for i in x_multiple]
      hasil_prediksi.append(label_multiple)
  # print(hasil_prediksi)

  if z != 1:
    # Ganti nilai-nilai dalam sublist hasil prediksi dengan nilai yang paling umum dari setiap sublist
    hasil_prediksi_modified = []
    for sublist in hasil_prediksi:
      count = Counter(sublist)
      most_common = count.most_common(1)[0][0]
      hasil_prediksi_modified.append(most_common)
    hasil_prediksi = hasil_prediksi_modified

  return hasil_prediksi


def jarak_euclidean(data,tetangga):
  x = data
  z = tetangga
  jarak_list = []

  # Iterasi melalui setiap baris dalam data frame
  for index, row in df_reg.iterrows():
      # Mengambil nilai fitur dari setiap baris dalam data frame
      x1, y1, z1 = row['A'], row['B'], row['C']  # Ganti dengan nama kolom sesuai data frame Anda

      # Menghitung jarak Euclidean antara data baru dan setiap baris dalam data frame
      jarak = math.sqrt((x[0] - x1)**2 + (x[1] - y1)**2 + (x[2] - z1)**2)

      # Menambahkan jarak ke dalam list jarak_list
      jarak_list.append(jarak)

  sorted_data = sorted(jarak_list)
  if z == 1:
    jarak_terkecil = sorted_data[0]
    indeks_jarakterkecil = jarak_list.index(jarak_terkecil)
    return indeks_jarakterkecil

  else:
    jarak_terkecil = sorted_data[:z]
    indeks_jarakterkecil = [jarak_list.index(jarak) for jarak in jarak_terkecil]
    return indeks_jarakterkecil
```

## Pembagian Data

```python
kolom_tertentu = ['A', 'B', 'C']  # Ganti dengan nama kolom yang diinginkan

x = df_reg[kolom_tertentu].iloc[:100]
y = df_reg['label'].iloc[:100]

x_tolist = x.values.tolist()
y_tolist = y.values.tolist()
```

##Prediksi Menggunakan data X

```python
data = x_tolist

#Usahakan Menggunakan Tetangga berjumlah Ganjil
tetangga = 1

y_pred = knn_predict(data,tetangga)
```

## Prediksi Menggunakan data Y

```python
akurasi = akurasi_score(y_tolist,y_pred)
akurasi
```
